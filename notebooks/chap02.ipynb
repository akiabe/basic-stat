```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../input/ch2_scores_em.csv", index_col="生徒番号")
df.head()
```

```python
scores = np.array(df["英語"][:10])
scores_df = pd.DataFrame(
    {"点数": scores},
    index=pd.Index(
        ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"], 
        name="生徒",
    )
)
scores_df
```

```python
print("平均")
print(np.mean(scores))
print(scores_df.mean())

print("-"*30)

print("中央値")
print(np.median(scores))
print(scores_df.median())

print("-"*30)

print("最頻値")
count = np.bincount(scores)
mode = np.argmax(count)
print(mode)

print(scores_df.mode())

print("-"*30)

print("偏差")
mean = np.mean(scores)
deviation = scores - mean
print(np.mean(deviation))

scores_df["偏差"] = deviation
print(scores_df)

print("-"*30)

print("分散")
variance = np.mean(np.square(deviation))
print(variance)

print(np.var(scores, ddof=0))

scores_df["分散"] = np.square(deviation)
print(scores_df)

print("-"*30)

print("標準偏差")
print(np.sqrt(np.var(scores, ddof=0)))
print(np.std(scores, ddof=0))

print("-"*30)

print("範囲と四分位範囲")
print(np.max(scores) - np.min(scores))

scores_Q1 = np.percentile(scores, 25)
scores_Q3 = np.percentile(scores, 75)
scores_IQR = scores_Q3 - scores_Q1
print(scores_IQR)

print("-"*30)

print("指標まとめ")
print(pd.Series(scores).describe())
```

```python
print("標準化")
scaling = (scores - np.mean(scores)) / np.std(scores)
print(np.mean(scaling), np.std(scaling, ddof=0))

scores_df["標準化"] = scaling
print(scores_df)
```

```python
print("度数分布")
english_scores = np.array(df["英語"])

print(pd.Series(english_scores).describe())
print("-"*30)

freq, _ = np.histogram(english_scores, bins=10, range=(0, 100))
freq_class = [f"{i}~{i+10}" for i in range(0, 100, 10)]
freq_dist_df = pd.DataFrame(
    {"度数": freq}, 
    index=pd.Index(
        freq_class, 
        name="階級",
    )
)
print(freq_dist_df)

print("-"*30)

class_value = [(i+(i+10))//2 for i in range(0, 100, 10)]
rel_freq = freq / freq.sum()
cum_rel_freq = np.cumsum(rel_freq)
freq_dist_df["階級値"] = class_value
freq_dist_df["相対速度"] = rel_freq
freq_dist_df["累積相対度数"] = cum_rel_freq
print(freq_dist_df)
```
